```
df = pd.read_excel(f'/content/drive/My Drive/{folder_name}/{file_name}.xlsx',header=1,usecols='A:B',keep_default_na=False, sheet_name='To be segregated')

df.head()

df.shape
```

```

s_ss_df = pd.read_excel(r'/content/drive/My Drive/Spain hospital dashboard/specialities & sub-specialities.xlsx',header=1,usecols='b:c',keep_default_na=False, sheet_name='Specialities')

s_ss_df['Speciality']=s_ss_df['Speciality'].str.lower().replace('\xa0',' ')

s_ss_df['Sub-Speciality']=s_ss_df['Sub-Speciality'].str.lower().replace('\xa0',' ')

s_ss_df.head()

s_ss_df.shape
```

```
def spec_nd_sub_spec(specs_str,s_ss_df):
    
    # [spec.lower().replace('\xa0',' ') for spec in list(set(s_ss_df['Speciality'].values))]
    
    specialities = list(set(s_ss_df['Speciality'].values))
    
    # [sspec.lower().replace('\xa0',' ') for sspec in list(set(s_ss_df['Sub-Speciality'].values))]
    
    sub_specialities = list(set(s_ss_df['Sub-Speciality'].values))
    
    spec,sub_spec=[],[]
    
    specs=specs_str.split('\n')
    
    for sp in specs:
        
        print('*',end='. ')
        
        sp=sp.lower().replace('\xa0',' ')
        
        #print(sp)
        
        if sp in specialities:
            
            spec.append(sp.title())

        else:

            if sp=='na':

              spec.append('NA')

              sub_spec.append('NA')

            else:

              spec.append('Other')

              sub_spec.append(sp.title())
            
        if sp in sub_specialities:
            
            sub_spec.append(sp.title())
            
            spec.append(s_ss_df[s_ss_df['Sub-Speciality']==sp]['Speciality'].iloc[0].title())

        else:

            if sp=='na':

              sub_spec.append('NA')
              
              spec.append('NA')

            else:

              sub_spec.append(sp.title())

              spec.append('Other')
            
    print('\n')
    
    #print(spec)
    
    if len(spec)==0:
        
        spec.append('NA')
            
    #print(sub_spec)
    
    if len(sub_spec)==0:
        
        sub_spec.append('NA')
    
    speciality='\n'.join(set(spec))
    
    sub_speciality='\n'.join(set(sub_spec))
    
    return pd.Series([speciality,sub_speciality])
```

```
df[['speciality','sub-speciality']]=df['Speciality(Total)'].apply(lambda x:spec_nd_sub_spec(x,s_ss_df))

df.head()
```

```
df.to_excel(f'/content/drive/My Drive/{folder_name}/{file_name}_processed.xlsx',index=False)
```
